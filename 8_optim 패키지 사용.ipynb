```python
import torch
from torch.autograd import Variable

N, D_in, H, D_out = 64, 1000, 100, 10

x = Variable(torch.randn(N,D_in))
y = Variable(torch.randn(N, D_out), requires_grad = False)

# nn 패키지를 사용하여 모델과 손실 함수 정의\
model = torch.nn.Sequential(torch.nn.Linear(D_in, H),
                            torch.nn.ReLU(),
                            torch.nn.Linear(H, D_out)
                           )
loss_fn = torch.nn.MSELoss(size_average=False)

# optim 패키지 사용
learning_rate = 1e-4
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate) # 첫번째 인자 : 갱신해야 하는 변수

# 순전파
for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred,y)
    print(t, loss.data)
    
    # 역전파 단계 전, Optimizer 객체를 사용하여 갱신할 변수들에 대한 모든 변화도를 0으로 만든다.
    # .backward()를 호출할 때마다 변화도가 버퍼(Buffer)에 누적되기 때문.
    # 우리의 목적은 누적이 아니라 새로 덮어쓰기
    optimizer.zero_grad()
    
    # 역전파 단계 : 모델의 매개변수에 대한 손실과 변화도 계산
    loss.backward()
    
    # Optimizer의 step 함수를 호출하면 매개변수가 갱신
    optimizer.step()
    
```
